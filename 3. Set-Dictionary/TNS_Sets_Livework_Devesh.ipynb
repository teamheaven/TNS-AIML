```python
#declaring a set
cars = {"honda","mahindra","ford","bmw"}
print("The brands gather for the event are: ",cars)
```

```python
#type of cars
cars = {"honda","mahindra","ford","bmw"}
print("The brands gather for the event are: ",cars)
print(type(cars))
```

```python
#add a element
cars = {"honda","mahindra","ford","bmw"}
cars.add("audi")
print("The brands gather for the event are: ",cars)
```

```python
#update
cars = {"honda","mahindra","ford","bmw"}
cars.update(["audi","tata","maruti"])
print("The brands gather for the event are: ",cars)
```

```python
number ={1,2,3,4,5}
number.add("dev")
print(number)
```

```python
#discard - don't throw and error
cars = {"honda","mahindra","ford","bmw"}
cars.discard("mercedes")
print("The brands gather for the event are: ",cars)
```

```python
#remove
cars = {"honda","mahindra","ford","bmw"}
cars.remove("ford")
print("The brands gather for the event are: ",cars)
```

```python
#pop
cars = {"honda","mahindra","ford","bmw"}
cars.pop()
print("The brands gather for the event are: ",cars)
```

```python
#clear
cars = {"honda","mahindra","ford","bmw"}
cars.clear()
print("The brands gather for the event are: ",cars)
```

```python
#difference
x = {"apple","banana","cherry"}
y = {"google","microsoft","apple"}
z = x.difference(y)
print(z)
```

```python
#union
x = {"apple","banana","cherry"}
y = {"google","microsoft","apple"}
z = x.union(y)
print(z)
```

```python
#intersection
x = {"apple","banana","cherry"}
y = {"google","microsoft","apple"}
z = x.intersection(y)
print(z)
```

```python
#symmetric difference
x = {"apple","banana","cherry"}
y = {"google","microsoft","apple"}
z = x.symmetric_difference(y)
print(z)
```

```python
#issubset
x = {"a","b","c"}
y = {"f","e","d","c","b","a"}
z = x.issubset(y)
print(z)
```

```python
#issuperset
x = {"f","e","d","c","b","a"}
y = {"a","b","c"}
z = x.issuperset(y)
print(z)
```
