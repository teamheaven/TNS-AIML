Python Dictionary -

Common Operations in Dictionary -
1. get
2. keys
3. values
4. item
5. update
6. pop
7. popitem
8. clear
9. copy
10. setdefault
11. del








```python
#declartion of dictionary
bike = {
    "user": "dev",
    "model_name": "triumph",
    "registered_no.": 123543,
    "color": "red"
}
print(tuple(bike))
```

```python
#get
bike = {
    "user": "dev",
    "model_name": "triumph",
    "registered_no.": 123543,
    "color": "red"
}
print(bike.get("user"))
```

```python
#value
bike = {
    "user": "dev",
    "model_name": "triumph",
    "registered_no.": 123543,
    "color": "red"
}
print(bike.values())
```

```python
#key
bike = {
    "user": "dev",
    "model_name": "triumph",
    "registered_no.": 123543,
    "color": "red"
}
a = bike.keys()
print(a)
```

```python
#item
bike = {
    "user": "dev",
    "model_name": "triumph",
    "registered_no.": 123543,
    "color": "red"
}
print(bike.items())
```

```python
#update
bike = {
    "user": "dev",
    "model_name": "triumph",
    "registered_no.": 123543,
    "color": "red"
}
car = {
    "user": "dev",
    "model_name": "tri",
    "registered_no.": 1243,
    "color": "blue"
}
a = bike.update(car)
print(a)
```

```python
#pop
bike = {
    "user": "dev",
    "model_name": "triumph",
    "registered_no.": 123543,
    "color": "red"
}
bike.pop("registered_no.")
print(bike)
```

```python
#popitem
bike = {
    "user": "dev",
    "model_name": "triumph",
    "registered_no.": 123543,
    "color": "red"
}
bike.popitem()
print(bike)
```

```python
#clear
bike = {
    "user": "dev",
    "model_name": "triumph",
    "registered_no.": 123543,
    "color": "red"
}
bike.clear()
print(bike)
```

```python
#pop
bike = {
    "user": "dev",
    "model_name": "triumph",
    "registered_no.": 123543,
    "color": "red"
}
bike.pop("registered_no.")
print(bike)
#
```

```python
#setdefault
bike = {
    "user": "dev",
    "model_name": "triumph",
    "registered_no.": 123543
}
bike.setdefault("color", "blue")
print(bike)
```

```python
#del
bike = {
    "user": "dev",
    "model_name": "triumph",
    "registered_no.": 123543,
    "color": "red"
}
del bike["registered_no."]
print(bike)
```
